```python
text_splitter = RecursiveCharacterTextSplitter(
    separators=["\n", "\. ", "\n\n", "\t"],
    keep_separator=True,
    chunk_size=10,
    chunk_overlap=10,
    strip_whitespace=True,
    is_separator_regex=False,
)

texts = text_splitter.create_documents([document])
texts = [doc.page_content for doc in texts]
for t in texts: 
    print(t)
    print('-----')
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")
doc = nlp(document)

phrases = [chunk.text for chunk in doc.sents]
# print(phrases)
for p in phrases:
    print(p)
    print('------')
```

```python
# Create phrase_data
phrase_data = [{
    'phrase': phrase
} for phrase in phrases]

print(len(phrase_data))
```

```python
# Calculate distances
from sentence_transformers import SentenceTransformer

# Download from the 🤗 Hub
model = SentenceTransformer("./chunking/embeddinggemma-300m")

phrase_encodings = model.encode_document([p.get('phrase') for p in phrase_data])
print(phrase_encodings[:3])

        

```

```python
# Loop through the list using an index
for i, phrase_dict in enumerate(phrase_data):
    phrase_dict['encoding'] = phrase_encodings[i]

phrase_data[:3]
```

```python
for i in range(0, len(phrase_data) - 1):
    distance = model.similarity(phrase_data[i]['encoding'], [phrase_data[i + 1]['encoding']])
    phrase_data[i]["distance"] = distance
    # print(distance)

phrase_data[:1]
```
